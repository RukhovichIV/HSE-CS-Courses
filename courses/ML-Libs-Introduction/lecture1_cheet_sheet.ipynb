```python
import time
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

# Структуры данных

## 1. Список (list)

Cheet link: https://wiki.python.org/moin/TimeComplexity

```python
# append complexity
n_iter = 1000
n_sizes = np.arange(10, 500, 10)
times = []
for n_size in tqdm(n_sizes):
    t0 = time.time()
    for _ in range(n_iter):
        x = []
        for i in range(n_size):
            x.append(0)
    times.append((time.time() - t0) / n_iter)
plt.plot(n_sizes, times, "o")
plt.grid()
```

```python
# getitem [] complexity
n_iter = 1000
n_sizes = np.arange(10, 500, 10)
times = []
for n_size in tqdm(n_sizes):
    x = [0] * n_size
    t0 = time.time()
    for i in range(n_iter):
        a = x[3*len(x)//4]
    times.append((time.time() - t0) / n_iter)
plt.plot(n_sizes, times, "o")
plt.grid()
```

```python
# pop(0), pop(-1) complexity
n_sizes = np.arange(10, 500, 10)
times = []
for n_size in tqdm(n_sizes):
    x = [0] * n_size
    n_iter = len(x)//2
    t0 = time.time()
    for i in range(n_iter):
        x.pop(0)
    times.append((time.time() - t0) / n_iter)
plt.plot(n_sizes, times, "o", label='pop(0)')

n_sizes = np.arange(10, 500, 10)
times = []
for n_size in tqdm(n_sizes):
    x = [0] * n_size
    n_iter = len(x)//2
    t0 = time.time()
    for i in range(n_iter):
        x.pop(-1)
    times.append((time.time() - t0) / n_iter)
plt.plot(n_sizes, times, "o", label='pop(-1)')
plt.grid()
plt.legend()
```

# 2. Map, set

```python
# add element complexity
n_iter = 1000
n_sizes = np.arange(10, 50000, 1000)
times = []
for n_size in tqdm(n_sizes):
    x = dict()
    idxs = np.random.randint(0, n_size * 100, size=n_size)
    t0 = time.time()
    for i in range(n_size):
        x[str(idxs[i])] = None
    times.append((time.time() - t0) / n_size)
plt.plot(n_sizes, times, "o")
plt.grid()
```

```python
# search element complexity
n_iter = 1000
n_sizes = np.arange(10, 50000, 1000)
times = []
for n_size in tqdm(n_sizes):
    idxs = np.arange(0, n_size, 1)
    x = {i: None for i in idxs}
    t0 = time.time()
    for i in range(n_size):
        a = i in x
    times.append((time.time() - t0) / n_size)
plt.plot(n_sizes, times, "o")

n_sizes = np.arange(10, 50000, 1000)
times = []
for n_size in tqdm(n_sizes):
    idxs = np.arange(0, n_size, 1)
    x = set({i for i in idxs})
    t0 = time.time()
    for i in range(n_size):
        i in x
    times.append((time.time() - t0) / n_size)
plt.plot(n_sizes, times, "o")
plt.grid()
```
