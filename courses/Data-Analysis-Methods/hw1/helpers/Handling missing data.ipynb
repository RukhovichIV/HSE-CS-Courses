```python
df_without_nan = df_def.copy()
df_without_nan = df_without_nan.dropna(axis=0)
```

```python
# df_without_nan
df_without_nan.info()
```

```python
from sklearn.linear_model import LinearRegression
df = df_def.copy()
for y in range(1, 7):
    x = [ i + 1 for i in range(6)]
    y_target = df_without_nan[f'{y}']
    x_train = df_without_nan.drop(f'{y}', axis=1)
    clf = LinearRegression().fit(x_train, y_target)
    for i in clf.predict(df_def[df_def[f'{y}'].isna()].drop(f'{y}', axis=1)):
        df = df.fillna(value={f"{y}": i}, limit=1)
        print({f"y": i})

clf.score(x_train, y_target)
clf.coef_
```

```python
from sklearn.linear_model import LogisticRegression
df = df_def.copy()
for y in range(1, 7):
    x = [ i + 1 for i in range(6)]
    y_target = df_without_nan[f'{y}']
    x_train = df_without_nan.drop(f'{y}', axis=1)
    clf = LogisticRegression(random_state=0).fit(x_train, y_target)
    for i in clf.predict(df_def[df_def[f'{y}'].isna()].drop(f'{y}', axis=1)):
        df = df.fillna(value={f"{y}": i}, limit=1)
        print({f"y": i})
clf.score(x_train, y_target)
clf.coef_
```
